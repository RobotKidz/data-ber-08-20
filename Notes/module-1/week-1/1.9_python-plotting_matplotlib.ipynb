```python
y1 = np.random.normal(loc=5, scale=3, size=50)
x1 = np.random.normal(loc=5, scale=3, size=50)

x2 = np.linspace(-5, 12.5, 30)
y2 = np.sin(x2)
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x1,
        y1,
        linestyle='-',
        marker='^',
        label='plot1',
       )
ax.plot(x2,
       y2,
        label='plot2'
       )
ax.set_xlabel('X- axis', fontsize=22)
ax.set_ylabel('Y- axis', fontsize=22)
ax.grid(axis='both')
ax.set_title('OUR FIRST PLOT', fontsize=50, pad=100)
ax.legend()
```
