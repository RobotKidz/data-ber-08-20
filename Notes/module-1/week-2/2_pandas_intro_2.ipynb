## 2. Sorting

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
# alternative to re-assigning you can use the inplace arg
# df.reset_index(drop=True, inplace=True)
```

```python
df[["kiwi_mass", "kiwi_mass", "orange_mass"]].sort_values("kiwi_mass")
```

```python
df.sort_values("kiwi_mass", ascending=False)
```

```python
df.sort_values(["kiwi_density", "kiwi_mass"])
```

```python
df = df.sort_values("kiwi_mass", ascending=False)
```

```python
df = df.reset_index(drop=True)
```

```python
# equivelent to the previous to cells - but done in one step. 
#Relies on the fact that these methods accept and return dataframes
df = (df
     .sort_values("orange_mass")
     .reset_index(drop=True))
```
