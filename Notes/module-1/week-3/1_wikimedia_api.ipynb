```python
def wiki_to_df(url):
    headers = {"accept":"application/json"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        wikistats = response.json()["items"]
        df = pd.DataFrame(wikistats)
        return df
    else:
        print("No success!")
```

Get data on a more granular level:
Wikimedia stats API offers more options. That is, pageview by:
- day
- access type
- agent type

## 1. access type

```python
url_desktop = "https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/en.wikipedia.org/desktop/all-agents/daily/20190101/20191231"
url_ma = "https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/en.wikipedia.org/mobile-app/all-agents/daily/20190101/20191231"
url_mw = "https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/en.wikipedia.org/mobile-web/all-agents/daily/20190101/20191231"
```

```python
wiki_to_df(url_desktop)
```

```python
access_types = ["desktop", "mobile-app", "mobile-web"]
wiki = pd.DataFrame(columns=df.columns)

for at in access_types:
    endpoint = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/en.wikipedia.org/{at}/all-agents/daily/20190101/20191231"
    wiki = pd.concat([wiki, wiki_to_df(endpoint)])
```

```python
wiki["access"].value_counts()
```

## 2. by agent type

```python
agent_types = ["user", "spider"]

wiki = pd.DataFrame(columns=df.columns)

for at in access_types:
    for agt in agent_types:
        endpoint = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/en.wikipedia.org/{at}/{agt}/daily/20190101/20191231"
        wiki = pd.concat([wiki, wiki_to_df(endpoint)])
```
